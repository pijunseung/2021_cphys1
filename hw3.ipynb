```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Leibniz Pi formula
$$\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}$$
이 식은 라이프니츠의 파이 공식이다.

1. 다음의 식을 구현하는 함수를 만들어라.
$$\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}$$
2. 이 식을 이용하여 아래처럼 그림을 그려라.

```python
def le(n):
    i, f = 1, 0
    while (i<=n):
        f = f + 1/((((-1)**(i-1))*(2*i-1)))
        i = i+1
    return f

x=np.arange(1,101,1)
y=[le(n) for n in x]
a=np.pi/4*x/x
plt.figure(figsize=(8, 8))
plt.xlabel('n', fontsize = 15)
plt.plot(x, y, label = r'Leibniz Formula for $\pi$')
plt.plot(x, a, label = r'$\pi/4$')
plt.legend()
```

## 2. Logistic Map
로지스틱 맵은 다음처럼 주어진다.$$ f_{n+1} = r f_n (1- f_n). $$

1. $f_n$을 구하는 함수를 작성하여라.
2. 작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.
  * $r=0.5$, $f_0$=0.5
  * $r=0.95$, $f_0$=0.5
  * $r=2$, $f_0$=0.5
  * $r=3.2$, $f_0$=0.5
  * $r=3.8$, $f_0$=0.5
  * $r=3.8$, $f_0$=0.501

```python
def logistic(r,f):
    g=[]
    for n in range(1,101):
        f=r*f*(1-f)
        g.append(f)
    return(g)
plt.figure(figsize=(8, 8))
plt.plot(logistic(0.5, 0.5), label = '$r=0.5$, $f_0$=0.5')
plt.plot(logistic(0.95, 0.5), label = '$r=0.95$, $f_0$=0.5')
plt.plot(logistic(2, 0.5), label = '$r=2$, $f_0$=0.5')
plt.plot(logistic(3.2, 0.5), label = '$r=3.2$, $f_0$=0.5')
plt.plot(logistic(3.8, 0.5), label = '$r=3.8$, $f_0$=0.5')
plt.plot(logistic(3.8, 0.501), label = '$r=3.8$, $f_0$=0.501')
plt.xlabel('n', fontsize =15)
plt.ylabel('$f_n$', fontsize = 15)
plt.legend(loc='best', bbox_to_anchor=(1, 0.5))
```

## 3. sinc function
1. 싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.$$ sinc(x)  = \frac{\sin x}{x}. $$
2. 싱크 함수를 테일러 전개하면 다음과 같다.$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$위의 식의 우변을 $n$을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.

```python
def facto(n):
    f = 1.0
    for i in range(1,2*n+2):
        f *= i
    return f

plt.figure(figsize=(10, 10))
x=np.arange(1,101,0.01)
plt.subplot(221)
plt.plot(x,np.sin(x)/x, label = 'sin(x)/x')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(222)
plt.plot(x, (((-1)**0)*(x**(2*0)))/facto(0), label = 'n=0')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(223)
plt.plot(x, (((-1)**1)*(x**(2*1)))/facto(1), label = 'n=1')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))

plt.subplot(224)
plt.plot(x, (((-1)**2)*(x**(2*2)))/facto(2), label = 'n=2')
plt.legend(loc='lower right', bbox_to_anchor=(1, 0.5))
```
